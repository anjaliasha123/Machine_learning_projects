```
import numpy as np
import matplotlib.pyplot as plt
```

```
K = 3
D = 2
N = 300
```

```
#creating sample data using mean and standard deviation
mu1 = np.array([0,0])
mu2 = np.array([5,5])
mu3 = np.array([0,5])

X = np.zeros((N,D))
X[:100,:] = np.random.randn(100,D) + mu1
X[100:200,:] = np.random.randn(100,D) + mu2
X[200:300,:] = np.random.randn(100,D) + mu3
Y = np.array([0]*100 + [1]*100 + [2]*100)
```

```
#making scatter plot of the data
plt.scatter(X[:,0],X[:,1],c=Y)
plt.xlabel('feature-1')
plt.ylabel('feature-2')
plt.show()
```

```
#obtaining centroids
means = np.zeros((K,D))
means[0,:] = X[Y==0].mean(axis=0)
means[1,:] = X[Y==1].mean(axis=0)
means[2,:] = X[Y==2].mean(axis=0)
```

```
#making scatter plot of the data
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(means[:,0],means[:,1],c='red',marker='*',s=200)
plt.xlabel('feature-1')
plt.ylabel('feature-2')
plt.show()
```

```
#test dataset clustering
testX = np.zeros((N,D))
testX[:100,:] = np.random.randn(100,D) + mu1 
testX[100:200,:] = np.random.randn(100,D) + mu2
testX[200:300,:] = np.random.randn(100,D) + mu3
```

```
#making scatter plot of the data
plt.scatter(testX[:,0],testX[:,1])
plt.xlabel('feature-1')
plt.ylabel('feature-2')
plt.show()
```

```
Y = np.zeros(N)
```

```
for n in range(N):
  closest_k = -1
  min_dist = float('inf')
  for k in range(K):
    d = ((testX[n,0]-means[k,0])**2 + (testX[n,1]-means[k,1])**2)**0.5
    if d < min_dist:
      min_dist = d
      closest_k = k
  Y[n] = closest_k
```

```
#making scatter plot of the data
plt.scatter(testX[:,0],testX[:,1],c=Y)
plt.scatter(means[:,0],means[:,1],c='red',marker='*',s=200)
plt.xlabel('feature-1')
plt.ylabel('feature-2')
plt.show()
```
